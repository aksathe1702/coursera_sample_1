```python
#4
import requests
from bs4 import BeautifulSoup
#4
import pandas as pd

def extract_gamestop_revenue():
  """Extracts GameStop's quarterly revenue data from Macrotrends."""

  url = 'https://www.macrotrends.net/stocks/charts/GME/gamestop/revenue'
  headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.0.0 Safari/537.36'}

  response = requests.get(url, headers=headers)
  soup = BeautifulSoup(response.content, 'html.parser')

  table = soup.find('table', {'class': 'table'})
  table_body = table.find('tbody')

  data = []
  for row in table_body.find_all('tr'):
    cols = row.find_all('td')
    if len(cols) >= 2:
      date = cols[0].text.strip()
      revenue = cols[1].text.strip().replace('$', '').replace(',', '')
      data.append([date, revenue])

  df = pd.DataFrame(data, columns=['Date', 'Revenue'])
  df['Revenue'] = pd.to_numeric(df['Revenue'])
  return df

if __name__ == '__main__':
  gamestop_revenue = extract_gamestop_revenue()
  print(gamestop_revenue)
```

```python
#5
import yfinance as yf
import pandas as pd
from alpha_vantage.fundamentaldata import FundamentalData
import plotly.express as px

# Extracting Tesla Stock Data
tesla = yf.Ticker("TSLA")
tesla_data = tesla.history(period="max")
tesla_data.reset_index(inplace=True)

# Extracting Tesla Revenue Data
api_key = 'YOUR_API_KEY'  # Replace with your actual Alpha Vantage API key
fd = FundamentalData(key=api_key, output_format='pandas')
revenue_data, meta_data = fd.get_income_statement_annual(symbol='TSLA')

# Merging Data (Assuming 'Date' column is present in revenue data)
merged_data = pd.merge(tesla_data, revenue_data[['fiscalYear', 'totalRevenue']], how='left', on='Date')

# Creating the Dashboard

# Stock Price vs. Date
fig1 = px.line(merged_data, x='Date', y='Close', title='Tesla Stock Price Over Time')

# Revenue vs. Year
fig2 = px.bar(merged_data, x='fiscalYear', y='totalRevenue', title='Tesla Annual Revenue')

# Layout (Optional - Adjust as needed)
fig1.update_layout(margin={'t': 40, 'b': 40, 'l': 20, 'r': 20}, width=700, height=500)
fig2.update_layout(margin={'t': 40, 'b': 40, 'l': 20, 'r': 20}, width=700, height=500)

# Display the Dashboard (using grid layout)
from plotly.subplots import make_subplots

rows = 1
cols = 2
fig = make_subplots(rows=rows, cols=cols, specs=[[{}, {}]])
fig.add_trace(fig1, row=1, col=1)
fig.add_trace(fig2, row=1, col=2)
fig.update_layout(title='Tesla Stock & Revenue Dashboard')
fig.show()
```

```python
#6
import yfinance as yf
import pandas as pd
import requests
from bs4 import BeautifulSoup
import plotly.express as px

# Extracting GameStop Stock Data
gamestop = yf.Ticker("GME")
gamestop_data = gamestop.history(period="max")
gamestop_data.reset_index(inplace=True)

# Extracting GameStop Revenue Data (using web scraping)
def extract_gamestop_revenue():
  """Extracts GameStop's quarterly revenue data from Macrotrends."""

  url = 'https://www.macrotrends.net/stocks/charts/GME/gamestop/revenue'
  headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.0.0 Safari/537.36'}

  response = requests.get(url, headers=headers)
  soup = BeautifulSoup(response.content, 'html.parser')

  table = soup.find('table', {'class': 'table'})
  table_body = table.find('tbody')

  data = []
  for row in table_body.find_all('tr'):
    cols = row.find_all('td')
    if len(cols) >= 2:
      date = cols[0].text.strip()
      revenue = cols[1].text.strip().replace('$', '').replace(',', '')
      data.append([date, revenue])

  df = pd.DataFrame(data, columns=['Date', 'Revenue'])
  df['Revenue'] = pd.to_numeric(df['Revenue'])
  return df

gamestop_revenue = extract_gamestop_revenue()

# Merge Data (Assuming date format is consistent)
merged_data = pd.merge(gamestop_data, gamestop_revenue, on='Date', how='left')

# Create the Dashboard

# Stock Price vs. Date
fig1 = px.line(merged_data, x='Date', y='Close', title='GameStop Stock Price Over Time')

# Revenue vs. Date
fig2 = px.bar(merged_data, x='Date', y='Revenue', title='GameStop Quarterly Revenue')

# Layout (Optional - Adjust as needed)
fig1.update_layout(margin={'t': 40, 'b': 40, 'l': 20, 'r': 20}, width=700, height=500)
fig2.update_layout(margin={'t': 40, 'b': 40, 'l': 20, 'r': 20}, width=700, height=500)

# Display the Dashboard (using grid layout)
from plotly.subplots import make_subplots

rows = 1
cols = 2
fig = make_subplots(rows=rows, cols=cols, specs=[[{}, {}]])
fig.add_trace(fig1, row=1, col=1)
fig.add_trace(fig2, row=1, col=2)
fig.update_layout(title='GameStop Stock & Revenue Dashboard')
fig.show()
```
